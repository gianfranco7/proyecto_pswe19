```python
import os
import csv
import numpy as np
import pandas as pd
import skfuzzy as fuzz
import matplotlib as mpl
from matplotlib import cm
from matplotlib import pyplot as plt
from sklearn.preprocessing import MinMaxScaler


#El csv debe estar en el mismo folder del archivo python
current_dir = os.getcwd()
filename = "incidencia.csv"

dataset_path = os.path.join(current_dir, filename)
df = pd.read_csv(dataset_path)

df.head()
```

# Normalizacion de Datos en inceidencias.csv
- Usando un MinMax Scaler de Sklearn


```python

# Seleccion de columnas numericas
numeric_columns = ['Madrugada', 'Mañana', 'Tarde', 'Noche', 'Total']

print(numeric_columns)

# Initialize the MinMaxScaler
scaler = MinMaxScaler()

# Normalizacion de Columnas
normalized_values = scaler.fit_transform(df[numeric_columns])

# Agregar columnas normalizadas al Dataframe
for i, col in enumerate(numeric_columns):
    df[f'{col}_norm'] = normalized_values[:, i]


df.sample(10)
```
